```python
import pandas as pd
from datetime import datetime
import math
import requests
import pandas as pd
import os
import re

df = pd.read_csv("data/CCI_List_of_Trades_2015-to-2023-05-13.csv")
df = df.iloc[2:]
df['date'] = df.apply(lambda x: x['Date/Time'][:10], axis=1)
df = df.drop_duplicates(subset='Trade #', keep='last').set_index('Trade #')

# extract data from df 
def extract(x):
    return x.split("_")

_, df['Close'], df['IV'], df['RV'], df['IV-RV'], df['Week'], df['Month'] = zip(*df['Signal'].map(extract))

# filter out rows with NaN values
df = df[df['RV'] != 'NaN']

# aggregate rows by dates
df = df.groupby('date').agg({'Profit USD': 'sum', 'Close': 'first', 'IV': 'first', 'RV': 'first', 'IV-RV': 'first', 'Week': 'first', 'Month': 'first'})

# get quarter data
df['Quarter'] = df.apply(lambda x: math.ceil(int(x['Month']) / 3), axis=1)

# convert datatype
for col in ['Close', 'IV', 'RV', 'IV-RV']:
    df[col] = df[col].astype('float')

for col in ['Week', 'Month']:
    df[col] = df[col].astype('int')

# classify rows to profitable or not
df['target'] = df.apply(lambda x: 1 if x['Profit USD'] > 0 else 0, axis=1)
```

```python
df["Yest_profits"] = df['Profit USD'].shift()
```

```python
# These are features computed with pinescript code on tradingview and exported 
# Implied Volatility Rank: daily implied volatility rank with a look back period of 252 trading days
# Implied Volatility Rank, or IV Rank & IVR for short, tells us whether implied volatility (IV) is high or low in a specific underlying based on the past year of IV data. For example, if XYZ has had an IV between 30 and 60 over the past year and IV is currently at 45, XYZ would have an IV rank of 50%.
# Rolling Hit Rate on a rolling basis with a look back period of 22 trading days - % of trades with positive profit
# Rolling Sharpe Ratio: average Sharpe ratio on a rolling basis with a look back period of 22 trading days - expected value over standard deviation of the daily P&L
df
```

```python
# Feature engineering
# for each variable compute: the daily difference (or first-order difference), the exponential moving average, the difference between the daily value and this moving average, and the standard deviation
# rolling basis using a window of 22 trading days

def variable_compute(col):
    df[col+"_diff"] = df[col] - df[col].shift()
    df[col+"_ema"] = df[col].ewm(span=22, adjust=False).mean()
    df[col+"_std"] = df[col].rolling(22).std()
    df[col+"_diff_ema"] = df[col+"_diff"] - df[col+"_ema"]

for col in ['Close', 'IV', 'RV', 'IV-RV', 'Yest_profits']:
    variable_compute(col)
    
# remove rows with NaN

df = df.dropna()
df = df.drop(columns=['Profit USD'])
df.to_csv("trades.csv")
```
